##Antecedentes Teóricos

1. **Regresión Logística**:
   - La regresión logística es un algoritmo de clasificación que se utiliza para problemas de clasificación binaria o multiclase.
   - Este método utiliza la función logística (también conocida como función sigmoide) para modelar la probabilidad de que una instancia pertenezca a una clase particular.
   - El objetivo principal de la regresión logística es encontrar los coeficientes óptimos que minimicen la función de pérdida, como la entropía cruzada o el error cuadrático medio.
   - La regresión logística es ampliamente utilizada en problemas de clasificación debido a su simplicidad y eficacia.

2. **K-Nearest Neighbors (KNN)**:
   - KNN es un algoritmo de clasificación y regresión que se basa en la proximidad de los datos.
   - Funciona calculando la distancia entre una instancia de prueba y los puntos de datos vecinos más cercanos en el espacio de características.
   - La decisión de clasificación se toma mediante votación (en problemas de clasificación) entre los K vecinos más cercanos.
   - KNN es un algoritmo no paramétrico, lo que significa que no asume una distribución específica para los datos.

3. **Máquinas de Soporte Vectorial (SVM)**:
   - Las SVM son un poderoso algoritmo de aprendizaje supervisado utilizado tanto para problemas de clasificación como de regresión.
   - SVM busca encontrar un hiperplano óptimo que maximice la separación entre clases en un espacio de características.
   - Puede manejar datos linealmente separables y no linealmente separables mediante el uso de funciones kernel.
   - SVM es conocido por su capacidad para lidiar con conjuntos de datos de alta dimensionalidad y su capacidad para generalizar bien en problemas de clasificación.

4. **Métricas de Desempeño**:
   - En la evaluación de algoritmos de clasificación, se utilizan métricas de desempeño como precisión, recall, F1-score y la matriz de confusión.
   - La precisión mide la proporción de predicciones correctas.
   - El recall (también llamado sensibilidad o tasa positiva verdadera) mide la proporción de verdaderos positivos identificados.
   - El F1-score es una medida que combina precisión y recall, proporcionando una métrica equilibrada.
   - La matriz de confusión muestra el número de verdaderos positivos, falsos positivos, verdaderos negativos y falsos negativos.


##Desarrollo

###Regresión Logistica

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv('/content/AnunciosRedesSociales.csv')
```

```python
X = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)
```

```python
scalerX = StandardScaler()
```

```python
x_train = scalerX.fit_transform(x_train)
x_test = scalerX.fit_transform(x_test)
```

```python
logreg = LogisticRegression(random_state=0)
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
pd.DataFrame(y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
X_set, y_set = x_train, y_train
X1, X2 = np. meshgrid(np.arange(start=X_set[:,0].min()-1,stop = X_set[:,0].max()+1,step=0.01),
                      np.arange(start=X_set[:,1].min()-1,stop = X_set[:,1].max()+1,step=0.01 ))

plt.contourf(X1,X2, logreg.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap = ListedColormap(('red','green')))

plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set==j,0], X_set[y_set==j,1],
              c= ListedColormap(('red', 'green'))(i),label = j)

plt.title('Clasificador train')
plt.xlabel('edad')
plt.ylabel('Salario estimado')
plt.legend()
plt.show()
```

```python
X_set, y_set = x_test, y_test
X1, X2 = np. meshgrid(np.arange(start=X_set[:,0].min()-1,stop = X_set[:,0].max()+1,step=0.01),
                      np.arange(start=X_set[:,1].min()-1,stop = X_set[:,1].max()+1,step=0.01 ))

plt.contourf(X1,X2, logreg.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap = ListedColormap(('red','green')))

plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set==j,0], X_set[y_set==j,1],
              c= ListedColormap(('red', 'green'))(i),label = j)

plt.title('Clasificador test')
plt.xlabel('edad')
plt.ylabel('Salario estimado')
plt.legend()
plt.show()
```

### Algoritmo Knn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv('/content/AnunciosRedesSociales.csv')
```
